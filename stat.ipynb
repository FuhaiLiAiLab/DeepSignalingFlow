## 1.1 Statistical Analysis

```python
import pandas as pd
kegg_df = pd.read_csv('./data-drugcomb-fi/raw_data/KEGG/full_kegg_pathway_list.csv')
display(kegg_df)
```

```python
# Get unique values and their counts
unique_values_counts = kegg_df['edge_type'].value_counts()
# Convert the result to a DataFrame
unique_values_counts_df = unique_values_counts.reset_index()
unique_values_counts_df.columns = ['Value', 'Count']

# Save the DataFrame to a CSV file
unique_values_counts_df.to_csv('./unique_values_counts.csv', index=False)
```
