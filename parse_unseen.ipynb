# 1. NCI ALMANAC

## 1.1 Mask some drugs

```python
import pandas as pd
import numpy as np
dataset = 'data-nci'
random_final_input_df = pd.read_csv('./' + dataset + '/filtered_data/random_final_dl_input.csv')
random_final_input_df
```

```python
# Combine sets of both columns to get unique drugs
unique_drug_list = sorted(list(set(random_final_input_df['Drug A']).union(set(random_final_input_df['Drug B']))))
np.random.shuffle(unique_drug_list)
print(unique_drug_list)
```

```python
# Splitting drug combinations into 80% and 20%
# Calculate the split index for 80% of the list
split_index = int(len(unique_drug_list) * 0.8)

# Split the list into 80% and 20%
drugs_train_percent = unique_drug_list[:split_index]
drugs_test_percent = unique_drug_list[split_index:]
print(len(drugs_train_percent), len(drugs_test_percent))
```

```python
# Filtering DataFrame by 80% cell lines
df_train_percent = random_final_input_df[random_final_input_df['Drug A'].isin(drugs_train_percent) | random_final_input_df['Drug B'].isin(drugs_train_percent)]
df_train_percent = df_train_percent.reset_index(drop=True)

# Filtering DataFrame by 20% cell lines
df_test_percent = random_final_input_df[random_final_input_df['Drug A'].isin(drugs_test_percent) & random_final_input_df['Drug B'].isin(drugs_test_percent)]
df_test_percent = df_test_percent.reset_index(drop=True)

display(df_train_percent)
display(df_test_percent)
df_train_percent.to_csv('./' + dataset + '/filtered_data/split_input_1.csv', index=False)
df_test_percent.to_csv('./' + dataset + '/filtered_data/split_input_2.csv', index=False)
```
